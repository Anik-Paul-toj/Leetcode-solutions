```python
# Build min heap(Heapify)
# Time complexity: O(n)
# Space complexity: O(1)

a = [-4,3,1,0,2,5,10,8,12,9]

import heapq
heapq.heapify(a)  # heapify the list to create a min heap ## heapify always makes a min heap
print(a)
```

```python
# Heap Push (Insert)
# Time complexity: O(log n)
# Space complexity: O(1)

heapq.heappush(a, 4)
print(a)
```

```python
# Heap pop (Remove)
# Time complexity: O(log n)
# Space complexity: O(1)

minn = heapq.heappop(a)
print(minn)
```

```python
# Heap Sort
# Time complexity: O(n log n)
# Space complexity: O(n)
# Note: O(1) space is possible via swapping elements in place, but it is not stable.

def heap_sort(arr):
    heapq.heapify(arr)
    print(arr)
    n = len(arr)
    sorted_arr = [0] * n
    
    for i in range(n):
        sorted_arr[i] = heapq.heappop(arr)
    return sorted_arr

arr = [5, 2, 9, 1, 5, 6]
sorted_arr = heap_sort(arr)
print(sorted_arr)
    
```

```python
# Max Heap
B = [-4,3,1,0,2,5,10,8,12,9]
n = len(B)

for i in range(n):
    B[i] = -B[i]
heapq.heapify(B)
print(B)
```

```python
largest = -heapq.heappop(B)
print(largest)
```

```python
minn = -heapq.heappop(B)
print(minn)
```

```python
heapq.heappush(B, -15) # want to insert 15 into the max heap so we push -15 into the min heap
print(B)
```
